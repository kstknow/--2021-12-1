# 다중회귀 - 보스톤 주택가격

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns= boston.feature_names)
df['PRICE'] = boston.target
df.head()
```

- 삼전도와 선형회귀선

```python
sns.regplot(x="CRIM", y ='PRICE', data = df)
plt.show()
```

```python
_, axs = plt.subplots(nrows =3, ncols=4, figsize=(16,12))  # fig를 사용하지 않아서 _사용  # _를 받지않으면 튜플로 받는다 # 튜플 언패킹 공부


for i, feature in enumerate(df.columns[1:-1]):                
    row, col = i//4, i%4
    sns.regplot(x=feature, y='PRICE', data= df, ax = axs[row][col])   # 13개 항목중에서 회귀선 근처에 몰려있는지
```

- feature 별 R square valuem

```python
from sklearn.linear_model import LinearRegression
```

```python
# CRIM
y = boston.target
X= df.CRIM.values.reshape(-1,1)
lr = LinearRegression()
lr.fit(X,y)
lr.score(X,y)
```

```python
df.CRIM.values.shape
```

```python
y = boston.target
for feature in df.columns[:-1]:
    X = df[feature].values.reshape(-1,1)
    lr = LinearRegression()
    lr.fit(X,y)
    score = lr.score(X,y)
    print(f'{feature}:\t{score:.4f}')
```

- 다중 선형 회귀(1차식으로만 )  -> 예측값을 활용

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    boston.data, boston.target, test_size= 0.1, random_state= 2021
)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 다중 선형회귀는 R squared Value가 무의미
lr.score(X_test, y_test)
```

```python
# 다중 선형회귀는 통계(변수)적 요소 중요
```

```python
# weight
lr.coef_
```

```python
# Bias
lr.intercept_
```

- 회귀식
<pre>
PRICE = -0.11 * CRIM = 0.053 * ZN + ... -0.55 * LSTAT +35.5(Bias)
<pre>

-  테스트 데이터

```python
X_test[0], y_test[0]
```

- 회귀식에 집어넣으면 예측값을 구할수 있다.

```python
# Weight와 Bias로 구한 예측값
```

```python
np.dot(X_test[0], lr.coef_) + lr.intercept_   # numpy array => 1차원 행렬이라면 각 자리 수끼리 곱하여 전부 더한다.
```

```python
# predict로 예측한 값
```

```python
lr.predict(X_test[:1]) # 데이터가 lr에 fit이나 predict은 [[],[],[]]-> 2차원 데이터가 와야됨, 그래서 reshape아니면 [:1] 해준다.
```

```python
# 10개의 데이터에 대한 예측값과 실제값
for i in range(10):
    pred1 = np.dot(X_test[0], lr.coef_) + lr.intercept_   # 중요 ^y = w.x + b -> w = lr.coef_ , .= np.dot, b -> lr.intercept_
    # pred2 = lr.predict(X_test[i].reshape(1,-1))
    # pred2 = lr.predict([X_test[i]])
    pred2 = lr.predict(X_test[i:i+1])
    print(f'실제값 : {y_test[i]}, 예측값1 :{pred1 : .2f}, 예측값2 : {pred2[0]:.2f}')
```
